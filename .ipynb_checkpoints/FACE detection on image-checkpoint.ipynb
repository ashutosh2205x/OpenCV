```python
# Main features 
#  Line features XnY, Edge features XnY, Four rectangle features of both white and black pixels
# on computation it would roughly provide the mean of both BnW pixels lying b/w 0-1, which will give also shades of gray
# Sum(whitePX)->Mean(Wpx) ; Sum(Bpx)-> mean(BPx); DELTA= M(W)- M(B) ; the delta close to 1, better the feature
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

```

```python
nadia = cv2.imread('./ASSETS/nadia.jpeg',0)
denis = cv2.imread('./ASSETS/denis.jpeg',0)
solvay = cv2.imread('./ASSETS/solvay.jpeg',0)

# plt.imshow(solvay, cmap='gray')
cascpath = ('/Users/mohitkumar/ashutosh-react-native/MY STUFF/OpenCV/ASSETS/haarcascade_frontalface_default.xml')

face_cascade = cv2.CascadeClassifier(cascpath)

def detect_face(img):
    face_img = img.copy()
#     x1,x2 y1,y2 positions of detected faces
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for(x,y,x1,y1) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+x1, y+y1),(255,255,255),10)
    return face_img


# 
def adj_detect_face(img):
    face_img = img.copy()
#     x1,x2 y1,y2 positions of detected faces
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor = 5, minNeighbors=2)
    for(x,y,x1,y1) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+x1, y+y1),(255,255,255),10)
    return face_img


result1 = detect_face(solvay)
result2 = detect_face(solvay)

# plt.imshow(result1, cmap='gray')
# print('/n')
plt.imshow(result2, cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('./ASSETS/haarcascade_eye.xml')

def detect_eyes(img):
    face_img = img.copy()
#     x1,x2 y1,y2 positions of detected eye
    eye_rects = eye_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors=10)
    
    for(x,y,x1,y1) in eye_rects:
        cv2.rectangle(face_img,(x,y),(x+x1, y+y1),(255,255,255),10)
    return face_img

result = detect_eyes(nadia)
plt.imshow(result, cmap ='gray')
```

```python
result = detect_eyes(denis)
plt.imshow(result, cmap ='gray')
```
