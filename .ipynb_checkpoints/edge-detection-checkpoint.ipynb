```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('./ASSETS/dog_face.jpg')

edges = cv2.Canny(image= img,threshold1=127, threshold2= 127)

plt.imshow(edges)
# TOO MUCH NOISE IN THE PICTURE
```

```python
#remove the noise calc. the median pixel value
med_val =np.median(img)
med_val

#set between 0 and 70% of the med_val, whichever is bigger
lower_val = int(max(0,0.7*med_val)) 

#set between full 255 and 130% of the med_val, whichever is smaller
upper_val = int(min(255,1.3*med_val))  
```

```python
edges = cv2.Canny(image=img,threshold1=lower_val,threshold2=upper_val+200)
plt.imshow(edges)       
```

```python
blurred = cv2.blur(img, ksize=(5,5))
edges = cv2.Canny(image=blurred,threshold1=lower_val,threshold2=upper_val)
plt.imshow(edges)   
```
